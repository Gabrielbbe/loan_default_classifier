# Testando normalidade

```python
def test_mean_difference(arra1, arra2):
    """
    Aplica o teste shapiro wilk para testar normalidade
    fazendo uma subamostragem dos dados pois o teste shapiro wilk
    test t consideram que o tamanho da amostra é pequeno
    além disso como temos um tamanho de amostra maior do que o recomendado
    para aplicar o shapiro wilk, podemos aplicar os testes de hipótese mais
    vezes e verificar o que acontece na maioria das vezes e ter um resultado
    final mais confiável
    """
    # Escolhe 5000 rows aleatoriamente dos dados originais
    # cada momento usa uma semente diferente então faz
    arr1, arr2 = np.random.choice(arra1,300), np.random.choice(arra2,300)

    _, pval1 = shapiro(arr1)
    _, pval2 = shapiro(arr2)
    
    alpha = 0.05  # nível de significância, nesse caso 5%
    
    if pval1 > alpha and pval2 > alpha:
        # Ambos os grupos são normalmente distribuídos, então aplica o teste t.
        _, pval_ttest = ttest_ind(arr1, arr2)
        test_result = "T-test"
    else:
        # pelo menos um dos grupos não são normais, aplica o Mann-Whitney U test
        _, pval_mannwhitneyu = mannwhitneyu(arr1, arr2, alternative='two-sided')
        test_result = "Teste U Mann-Whitney"
    
    # Determina se as médias são diferentes baseado nos valores
    # nos p-valores dos testes.
    if (pval1 > alpha and pval2 > alpha and pval_ttest < alpha) or \
       (pval1 <= alpha or pval2 <= alpha and pval_mannwhitneyu < alpha):
        means_different = True
    else:
        means_different = False
    
    return test_result, means_different
```

```python
porc__dif_means = {
    'person_age':[],
    'person_income':[],
    'person_emp_length':[],
    'loan_amnt':[],
    'loan_int_rate':[],
    'loan_percent_income':[],
    'cb_person_cred_hist_length':[]
}
columns = ['person_age','person_income',
            'person_emp_length','loan_amnt','loan_int_rate','loan_percent_income','cb_person_cred_hist_length']
inadimplentes = df_teste.loc[ df_teste['loan_status']==1 ]
n_inadimplentes = df_teste.loc[ df_teste['loan_status']==0 ]

for i in range(50):

    for col in columns:
        
        #print(col)
        arr1 = inadimplentes[col]
        arr2 = n_inadimplentes[col]

        result, different_means = test_mean_difference(arr1, arr2)
        porc__dif_means[col].append(different_means)
        # print(f"Statistical test used: {result}")
        # if different_means:
        #     print(col + " possue Médias que são estatisticamente diferentes. \n" )
        # else:
        #     print(col + " possue Médias que não são estatisticamente diferentes. \n")
```

##### Vamos fazer um teste de hipótese afim de verificar se a porcentagem de inadimplentes e não inadimplentes em cada categoria dentro das variáveis possuem diferença, assim indicando que uma categoria tende a ter mais ou menos inadimplentes.

```python
df = pd.read_csv('../dados/credit_risk_dataset.csv')
```

```python
from scipy.stats import chi2_contingency
soma_testes = {'cb_person_default_on_file':0, 'loan_grade':0, 'loan_intent':0,
               'person_home_ownership':0}
for i in range(50):
        
    for col in colunas_categorical:
        conj_teste = np.random.choice(df['Unnamed: 0'],300)
        cats = df.iloc[conj_teste]
        # Criar uma tabela de contingência
        contingency_table = pd.crosstab(cats[col], cats[ "loan_status"])

        # Realizar o teste do qui-quadrado
        chi2, p, _, expected = chi2_contingency(contingency_table)

        print("Valor de chi2:", chi2)
        print("Valor p:", p)

        if p < 0.05:
            print(f"Rejeitamos a hipótese nula. Há uma relação significativa entre '{col}' e 'loan_status'.")
            soma_testes[col] = soma_testes[col] + 1
        else:
            print(f"Não rejeitamos a hipótese nula. '{col}' e 'fraude' parecem ser independentes.")
```

##### Afim de verificar quais variáveis categóricas possuem maior correlação com a variável resposta vamos calcular o coeficiente V de Cramer de cada covariável.

```python
def table_cramer_v(df, categorical_var, category_list):
    
    table = pd.DataFrame( columns = category_list )
    table.loc[0,:] = 0

    for col in category_list:
        
        contingency_table = pd.crosstab(df[col], df[categorical_var])
        chi2, _, _, _ = chi2_contingency(contingency_table)
        n = contingency_table.sum().sum()
        phi2 = chi2 / n
        r, k = contingency_table.shape
        phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
        rcorr = r - ((r-1)**2)/(n-1)
        kcorr = k - ((k-1)**2)/(n-1)
        cramer_v = np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))
        
        table[col] = cramer_v
        
    return table
```

```python
var_interesse = 'loan_status'
tabela_cramer = table_cramer_v(df_teste, 'loan_status',colunas_categorical)
print( f'Coeficiente V de cramer em relação a {var_interesse}' )
tabela_cramer.T.sort_values(by = 0, ascending=False).style.bar(color='blue')
```

### Sobre o preprocessamento nas variáveis categóricas
#### Alguns autores argumentam que devemos considerar uma variável categórica ordinal como ordinal caso ela possua relação de ordinalidade com a variável resposta, e não em relação ao seu significado inerente, com isso em mente e com o nosso entendimento dos dados podemos aplicar uma transformação ordinal na variável de grau do empréstimo (loan_grade), na variável do tipo de residência do cliente (person_home_ownership) podemos aplicar a mesma transformação, e na variável que indica se a pessoa já foi inadimplente ou não (cb_person_default_on_file) que é uma variável dicotômica, na variável que indica a intenção do empréstimo (loan_intent) podemos aplicar one-hot-encoding onde transformamos cada categoria numa coluna dicotômica que indica a presença ou ausência da categoria, as variáveis loan_intent e loan_grade apresentam uma cardinalidade média, caso o modelo não performe muito bem podemos tentar reduzir o número de classes dentro dessa variável para melhorar a acurácia do modelo no momento da modelagem.
